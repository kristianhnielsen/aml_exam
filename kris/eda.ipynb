```python
for col in data.columns:
    print(f"{col}: \t{data[col].nunique()} unique values")
    print(data[col].unique())
    sns.countplot(x=col, data=data, legend=True)
    plt.show()
    print()
```

```python
for col in data.columns:
    churned = data[data["churn"] == "Yes"]
    not_churned = data[data["churn"] == "No"]
    sns.countplot(x=col, data=data, hue="churn")
    plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, drop="first")
data["gender"] = ohe.fit_transform(data[["gender"]])
```

```python
binary_cols = (
    data.select_dtypes(include=["object"])
    .nunique()[data.select_dtypes(include=["object"]).nunique() == 2]
    .index
)
binary_cols
```

```python
yes_no_map = {"Yes": 1, "No": 0}
for col in binary_cols:
    data[col] = data[col].map(yes_no_map)
```

```python
data.head()
```

```python
data.describe()
```

```python
corr_data = data.copy()
plt.figure(figsize=(25, 10))
corr = corr_data.apply(lambda x: pd.factorize(x)[0]).corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

ax = sns.heatmap(
    corr,
    mask=mask,
    xticklabels=corr.columns.tolist(),
    yticklabels=corr.columns.tolist(),
    annot=True,
    linewidths=0.2,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
)
plt.show()
```
