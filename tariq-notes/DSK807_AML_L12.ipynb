# **Image Processing**

![](https://www.theclickreader.com/wp-content/uploads/2020/08/color-channels-RGB.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load an image (JPEG file)
img = Image.open("./datasets/face.png")
img = img.convert("RGB")          # ensure it's in RGB format

# Convert to NumPy array
img_array = np.array(img)
print("Image shape:", img_array.shape)  # (height, width, 3)
```

```python
# Split the RGB channels
R = img_array[:, :, 0]
G = img_array[:, :, 1]
B = img_array[:, :, 2]

# Display all channels
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.imshow(img_array)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(R, cmap='Reds')
plt.title("Red Channel")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(G, cmap='Greens')
plt.title("Green Channel")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(B, cmap='Blues')
plt.title("Blue Channel")
plt.axis("off")

plt.tight_layout()
plt.show()
```

# **2. Convert RGB to Grayscale/BW**

### 2.1. Averaging RGB Values

1.   List item
2.   List item



```python
# Convert RGB to Grayscale using the average
gray_avg =  (R + G + B)/3
gray_avg = gray_avg.astype(np.uint8)

plt.imshow(gray_avg, cmap='gray')
plt.title("Grayscale Image (averaging)")
plt.axis("off")
plt.show()
```

### 2.2. Using the Luminosity Formula

```python
# Convert RGB to Grayscale using the luminosity formula
gray_lum = 0.299 * R + 0.587 * G + 0.114 * B
gray_lum = gray_lum.astype(np.uint8)

# 6. Show grayscale image
plt.imshow(gray_lum, cmap='gray')
plt.title("Grayscale Image (luminosity formula)")
plt.axis("off")
plt.show()
```

### 2.3. Convert into Black/White
(using a threshold)

```python
# Black and White

T = 128
bw = np.where(gray_lum > T, 255, 0).astype(np.uint8)

plt.imshow(bw, cmap='gray')
plt.title("Black/White Image")
plt.axis("off")
plt.show()
```

# 3. Subsampling
Subsampling keeps one pixel out of every `x` in both directions:

`
new_image = old_image[::x, ::x]
`

 reduces dimensions by `1/x` in height and width.

```python
# Display results
plt.figure(figsize=(16,4))
plt.subplot(1,4,1); plt.imshow(gray_lum, cmap='gray'); plt.title("Original Grayscale"); plt.axis('off')
plt.subplot(1,4,2); plt.imshow(gray_lum[::4,::4], cmap='gray'); plt.title("Subsampling by 1/2"); plt.axis('off')
plt.subplot(1,4,3); plt.imshow(gray_lum[::16,::16], cmap='gray'); plt.title("Subsampling by 1/16"); plt.axis('off')
plt.subplot(1,4,4); plt.imshow(gray_lum[::32,::32], cmap='gray'); plt.title("Subsampling by 1/32"); plt.axis('off')
plt.show()
```

# 4. Pooling
![](https://miro.medium.com/1*fXxDBsJ96FKEtMOa9vNgjA.gif)

Image Source: [towardsai.net](https://pub.towardsai.net/introduction-to-pooling-layers-in-cnn-dafe61eabe34)





```python
pool_size = 5

print("Original shape:", gray_lum.shape)
h, w = gray_lum.shape

new_h = h // pool_size
new_w = w // pool_size

mean_pooled_img = np.zeros((new_h, new_w)) # Mean Pooling
max_pooled_img = np.zeros((new_h, new_w)) # Max Pooling

# Perform Pooling
for i in range(new_h):
    for j in range(new_w):
        region = gray_lum[i*pool_size:(i+1)*pool_size,
                        j*pool_size:(j+1)*pool_size]
        mean_pooled_img[i, j] = np.mean(region)
        max_pooled_img[i, j] = np.max(region)

print("Pooled shape:", mean_pooled_img.shape)

# Display results
plt.figure(figsize=(16,4))
plt.subplot(1,4,1); plt.imshow(gray_lum, cmap='gray'); plt.title("Original Grayscale"); plt.axis('off')
plt.subplot(1,4,2); plt.imshow(mean_pooled_img, cmap='gray'); plt.title(f"Mean Pooling (x{1/pool_size})"); plt.axis('off')
plt.subplot(1,4,3); plt.imshow(max_pooled_img, cmap='gray'); plt.title(f"Max Pooling (x{1/pool_size})"); plt.axis('off')
plt.subplot(1,4,4); plt.imshow(gray_lum[::5, ::5], cmap='gray'); plt.title(f"[::5, ::5]"); plt.axis('off')
plt.show()
```

# Filtering

```python
gaussian_kernel = (1/16) * np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]
])

sharpen_kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
])

import cv2

gaussian_filtered = cv2.filter2D(gray_lum, -1, gaussian_kernel)
sharpen_filtered = cv2.filter2D(gray_lum, -1, sharpen_kernel)
# Display results
plt.figure(figsize=(16,4))
plt.subplot(1,4,1); plt.imshow(gray_lum, cmap='gray'); plt.title("Original Grayscale"); plt.axis('off')
plt.subplot(1,4,2); plt.imshow(gaussian_filtered, cmap='gray'); plt.title("Gaussian Blur Kernel"); plt.axis('off')
plt.subplot(1,4,3); plt.imshow(sharpen_filtered, cmap='gray'); plt.title(f"Sharpen Filter"); plt.axis('off')
plt.show()
```

```python
# Define useful filters
filters = {
"Identity": np.array([[0,0,0],[0,1,0],[0,0,0]]),
"Box Blur 3": (1/9) * np.ones((3,3)),
"Box Blur 30": (1/900) * np.ones((30,30)),
"Gaussian Blur": (1/16) * np.array([[1,2,1],[2,4,2],[1,2,1]]),
"Sharpen": np.array([[0,-1,0],[-1,5,-1],[0,-1,0]]),
"Edge Detection (Sobel X)": np.array([[-1,0,1],[-2,0,2],[-1,0,1]]), # Detects vertical edges (intensity change in X direction).
"Edge Detection (Sobel Y)": np.array([[-1,-2,-1],[0,0,0],[1,2,1]]), # Detects horizontal edges (intensity change in Y direction).
"Emboss": np.array([[-2,-1,0],[-1,1,1],[0,1,2]]), # Creates a 3D relief effect emphasizing contours.
"Edge Enhancement": np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]]), # Sharpens and emphasizes edges by boosting contrasts.
"Laplacian": np.array([[0,1,0],[1,-4,1],[0,1,0]]) # Highlights regions of rapid intensity change;
}

# Apply and display results
plt.figure(figsize=(14, 10))
for idx, (name, kernel) in enumerate(filters.items()):
  filtered = cv2.filter2D(gray_lum,-1, kernel)
  plt.subplot(4, 3, idx+1)
  plt.imshow(filtered, cmap='gray')
  plt.title(name)
  plt.axis('off')
plt.tight_layout()
plt.show()
```

# Convolutional Neural Networks (CNN)

```python
import tensorflow as tf
from matplotlib import pyplot as plt

# Load the data
data = tf.keras.datasets.cifar10.load_data()
# Split the data
(train_images, train_labels), (test_images, test_labels) = data
train_images, test_images = train_images / 255.0, test_images / 255.0
train_labels, test_labels = train_labels.reshape(-1), test_labels.reshape(-1)
print(train_images.shape, train_labels.shape, test_images.shape, test_labels.shape)

class_names = [
    'airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse',
    'ship', 'truck'
]



plt.figure(figsize=(8, 8))
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(32, 32, 3)),
    tf.keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    tf.keras.layers.Flatten(), # flatten before fully connected part
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax'),
    ])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'],
    )
model.summary()
```

```python
history = model.fit(train_images, train_labels,
                    validation_data=(test_images, test_labels), epochs=15,
                    verbose=1)
```

```python
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label = 'Test')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```

## Regularization
![](https://www.mdpi.com/applsci/applsci-12-07682/article_deploy/html/images/applsci-12-07682-g001.png)

Image Source: https://www.mdpi.com/2076-3417/12/15/7682

```python
model_adv = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(32, 32, 3)),
    tf.keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(), # batch normalization
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(), # batch normalization
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu'), # bigger
    tf.keras.layers.BatchNormalization(), # batch normalization
    tf.keras.layers.SpatialDropout2D(0.2), # dropout, but for filters instead of nodes
    tf.keras.layers.Flatten(), # flatten before fully connected part
    tf.keras.layers.Dense(128, activation='relu'), # bigger
    tf.keras.layers.BatchNormalization(), # batch normalization
    tf.keras.layers.Dropout(0.2), # normal dropout
    tf.keras.layers.Dense(10, activation='softmax'),
    ])

model_adv.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'],
    )
model_adv.summary()
```

```python
history_adv = model_adv.fit(train_images, train_labels,
                            validation_data=(test_images, test_labels), epochs=15, verbose=1)
```

```python
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label = 'Test')
plt.plot(history_adv.history['accuracy'], label = 'Train (adv. model)')
plt.plot(history_adv.history['val_accuracy'], label = 'Test (adv. model)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```

```python
model.evaluate(test_images,  test_labels)
model_adv.evaluate(test_images,  test_labels);
```
